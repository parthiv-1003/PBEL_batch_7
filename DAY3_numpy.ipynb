```python
#create a guessing game where uesr will enter a number you have tell number is big or small
#loop user guess another number untillit win

num = np.random.randint(0,10)
user = int(input("enter a number :"))
while(num != user):
    user = int(input("enter a number :"))
    if (num == user):
        print("you won")
    elif(num>user):
        print("guess a large number")
    else:
        print("guess a small number ")
```
